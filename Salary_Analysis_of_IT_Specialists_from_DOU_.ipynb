7. Transform the 'Timestamp' column to the data type `datetime64[ns]`

```python
import datetime
```

```python

def get_time(date):
  correct_date = None
  if len(date) == 19:
    correct_date = datetime.datetime.strptime(date, "%m.%d.%Y %H:%M:%S")
  elif len(date) == 16:
    correct_date = datetime.datetime.strptime(date, "%m.%d.%Y %H:%M")
  return correct_date

```

```python
date_string = "11.30.2022 10:26:37"
date_str = "01.09.2023 11:38"
print(len(date_string), len(date_str))

```

```python
print(get_time(date_str))
```

```python
df['Timestamp'] = df.Timestamp.apply(get_time)
```

11. Sorting the English proficiency levels

```python
df['Знання англійської мови'] = pd.Categorical(df['Знання англійської мови'],
                      categories=['Elementary','Pre-Intermediate','Intermediate','Upper-Intermediate','Advanced'],
                      ordered=True)

df['Ваш тайтл'] = pd.Categorical(df['Ваш тайтл'],
                      categories=['Intern/Trainee', 'Junior', 'Немає тайтлу', 'Middle', 'Senior', 'Tech Lead', 'Team Lead', 'Architect', 'Head', 'Manager'],
                      ordered=True)
```

```python
df.dtypes
```

12. Creating a pivot table where the columns are 'Your Job Title', the rows are 'English Proficiency Level', and the values represent the median salary

```python
pivot_salary_englevel_title = pd.pivot_table(df, index='Ваш тайтл', columns='Знання англійської мови', values='Зарплата у $$$ за місяць, лише ставка після сплати податків', aggfunc='median')
pivot_salary_englevel_title = pivot_salary_englevel_title.style.format('${:.0f}')
pivot_salary_englevel_title = pivot_salary_englevel_title.background_gradient(axis=None, cmap='YlOrRd')

pivot_salary_englevel_title
```

Salary increases with the level of job title. More experienced professionals (such as Senior, Tech Lead, Architect) earn significantly higher salaries compared to lower titles (Intern/Trainee, Junior).

An improvement in the level of English proficiency can increase the salary by several thousand dollars, especially for Middle, Senior, and Tech Lead roles. However, for some lower job titles (like Intern/Trainee), the importance of English proficiency might be less, but it still contributes to a higher salary
